```python
A=pd.DataFrame(np.random.randint(0,20,(3,3)),columns=list('PRT'))
B=pd.DataFrame(np.random.randint(10,20,(3,3)),columns=list('PRT'))
print(A)
print(B)
```

```python
pd.concat([A,B],axis=1)
```

```python
pd.concat([A,B],axis=0)
```

```python
pd.concat([A,B],axis=1,ignore_index=True)
```

```python
#inner join or outer join 
# simple pass union 
# inner pass intersection 
# join_axes pass take column of df2
print(pd.concat([A,B]))
print("-------------")
print(pd.concat([A,B],join='inner'))
print("-------------")
print(pd.concat([A,B],join_axes=[B.columns]))
```

```python
import numpy as np
rng=np.random.RandomState(42)
df=pd.DataFrame({'key':['A','B','C','A','B','C'] , 'data1':range(6),'data2':rng.randint(0,10,6)},columns=['key','data1','data2'])
df
```

```python
#aggregate()=it is used to find single value => it takes list,function,dictionary
#filter()   =it is used to find those value which is in given condition =>it takes function,list,tuple
#transform()=it is used to overwrite the data =>it takes only function
```

```python
df.groupby('key').aggregate([min,np.median,max])
```

```python
 df.groupby('key').aggregate({'data1':min,'data2':max})
```

```python
#Filter
print(df.query('data2>3'))
print('\n=======================================\n')
df.query('data2>5')
```

```python
df.query('data1>3 and  data2>5')
```

```python
df.query('data1 + data2 >=5')
```

```python
df[(df.data1>=3) & (df.data2>5)] 
```

```python
df[(df['data1']+df['data2'])>=5]
```

```python
df.groupby('key').filter(lambda x:x['data2'].std()>2) #find standard deviation of data2 #lambda returns true or  false
```

```python
def filter1():
    filter1=lambda x:x['data2'].std()>2
    return filter1

df.groupby('key').filter(filter1())
```

```python
def filter2(x):
    return(x['data2'].std()>2)
    
    
    
df.groupby('key').filter(filter2)
```

# Transform

```python
print(df,'\n=============================================\n')
print(df.groupby('key').mean(),'\n=============================================\n')
print(df.groupby('key').transform(lambda x:x-x.mean()))
```
